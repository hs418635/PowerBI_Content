# Splitting training and testing dataset

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2)
```

```python
train.shape
```

```python
test.shape
```

### Checking indipendent variable 

```python
trainx=train.iloc[:,:-1]
trainx
```

### target variable of train data

```python
trainy=train.iloc[:,-1]
trainy
```

```python
testx=test.iloc[:,:-1]
testx
```

```python
testy=test.iloc[:,-1]
testy
```

# Performing Regression Model

# building Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(trainx,trainy)
pred=logit.predict(testx)
pred
```

### checking accuracy with logistic regression

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(testy,pred)
acc
```

# Performing classification Algorithms

# Building Knn Model

```python
from sklearn.neighbors import KNeighborsClassifier
Knn = KNeighborsClassifier()
Knn.fit(trainx,trainy)
pred_Knn=Knn.predict(testx)
pred_Knn
```

### Checking accuracy with Knn 

```python
from sklearn.metrics import accuracy_score
acc1=accuracy_score(testy,pred_Knn)
acc1
```

# Building Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(trainx,trainy)
pred_nb=nb.predict(testx)
pred_nb
```

### Checing accuracy with Naive Bayes 

```python
from sklearn.metrics import accuracy_score
acc2=accuracy_score(testy,pred_nb)
acc2
```

# Building Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dtr = DecisionTreeClassifier()
dtr.fit(trainx,trainy)
pred_dtr=dtr.predict(testx)
pred_dtr
```

### Checing accuracy with Decision Tree 

```python
from sklearn.metrics import accuracy_score
acc3=accuracy_score(testy,pred_dtr)
acc3
```

# Building RandomForest Model

```python
from sklearn.ensemble import RandomForestClassifier
rdm = RandomForestClassifier()
rdm.fit(trainx,trainy)
pred_rdm=rdm.predict(testx)
pred_rdm
```

## Checking accuracy with RandomForest

```python
from sklearn.metrics import accuracy_score
acc4 = accuracy_score(testy,pred_rdm)
acc4
```
